```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import sklearn.datasets
from sklearn.model_selection import train_test_split
import seaborn as sns
from xgboost import XGBRegressor
from sklearn import metrics
```

#### importing house price prediction dataset

```python
dataset = sklearn.datasets.fetch_california_housing()
```

```python
print(dataset)
```

```python
#loading the dataset 
houseprice_dataset = pd.DataFrame(dataset.data, columns = dataset.feature_names)
```

```python
houseprice_dataset
```

```python
#add the target(price) column to the dataframe
houseprice_dataset['Price'] = dataset.target
```

```python
houseprice_dataset.head()
```

```python
houseprice_dataset.shape
```

```python
houseprice_dataset.isnull().sum()
```

```python
houseprice_dataset.describe()
```

### EXPLORATORY DATA ANALYSIS
## correlation

```python
correlation = houseprice_dataset.corr()
```

##### constructing heat heat map


```python
plt.figure(figsize = (10,10))
sns.heatmap(correlation, cbar = True, fmt = '.3f', annot = True, annot_kws={'size':8}, cmap= 'Blues')
```

```python
#splitting the data and target(price) to get the prediction
X = houseprice_dataset.drop(['Price'], axis = 1)
Y = houseprice_dataset['Price']
```

```python
print(X)
print(Y)
```

```python
#splitting the data into training data and test data
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size = 0.2, random_state = 2)
```

```python
print(X.shape, X_train.shape, X_test.shape)
```

### model training

#### XGBooest regressor


```python
model = XGBRegressor()
```

```python
#training the model with X_train
model.fit(X_train, Y_train)
```

##### evaluation

```python
#prediction on training data
training_data_prediction = model.predict(X_train)
```

```python
print(training_data_prediction)
```

```python
# R squared error
score_1 = metrics.r2_score(Y_train, training_data_prediction)
```

```python
# Mean absolute error
score_2 = metrics.mean_absolute_error(Y_train,training_data_prediction)
```

```python
print("R squred error(training data):", score_1)
print("mean absolute error(training data):", score_2)
```

```python
test_data_prediction = model.predict(X_test)
```

```python
print(test_data_prediction)
```

```python
# R squared error
score_3 = metrics.r2_score(Y_test, test_data_prediction)
```

```python
# Mean absolute error
score_4 = metrics.mean_absolute_error(Y_test,test_data_prediction)
```

```python
print("R squred error(test data):", score_3)
print("mean absolute error(test data):", score_4)
```

```python
#visualing the actual prices and predicted prices
plt.scatter(Y_train, training_data_prediction)
plt.xlabel("Actual prices")
plt.ylabel("Predicted prices")
plt.title("actual price v/s predicted price")
plt.show()
```
